# Create your own plugin

We will use here the software cookiecutter to create a plugin with good defaults. First, if not yet there we need to install cookiecutter.

1. Activate your environment
2. install cookiecutter:\
   ```pip install cookiecutter```

Once this is done, follow the steps explained [here](https://kevinyamauchi.github.io/open-image-data/tutorials/neubias_plugins/make_a_simple_plugin.html#using-the-cookiecutter-template-to-create-your-plugin-directory).

````python
import napari
from napari.utils import nbscreenshot
````

````python
viewer = napari.Viewer()
````

````python
from skimage import io

spots_url = 'https://raw.githubusercontent.com/kevinyamauchi/napari-spot-detection-tutorial/main/data/spots_cropped.tif'
spots = io.imread(spots_url)
````

````python
viewer.add_image(spots)
````

## Adding your plugin programmatically

When you develop a plugin, it's annoying to continuously have to click in the interface to make the plugin appear. We can do it programmatically, just as previously:

````python
from napari_dslcourse._widget import ExampleQWidget
````

````python
viewer = napari.Viewer()
mywidget = ExampleQWidget(viewer)
viewer.window.add_dock_widget(mywidget)
````

````python
nbscreenshot(viewer)
````

## Hot reloading

When developing a plugin it is also annoying to have to restart and reload packages at every change. To avoid that one can use hot reloading, i.e. changes done in the code (if installed locally with ```pip install -e .``` will be taken into account immediately in the notebook. What remains necessary is to **create a new napari Viewer** as napari is not capable of hot reloading the plugin.

To use hot reloading, execute the two lines below at the very top of your notebook, before any import:

````python
%load_ext autoreload
%autoreload 2
````
